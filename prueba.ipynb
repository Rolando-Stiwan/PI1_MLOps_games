```python
import pandas as pd
import gzip
import json
import ast
```

```python
# Ruta al archivo comprimido que contiene el JSON en formato Gzip
ruta_archivo_comprimido = 'steam_games.json.gz'

# Nombre del archivo JSON dentro del archivo comprimido
nombre_archivo_json = 'output_steam_games.json'

# Abre el archivo comprimido en modo lectura
with gzip.open(ruta_archivo_comprimido, 'rb') as archivo_gzip:
    # Lee el contenido del archivo Gzip y carga los datos en un DataFrame de Pandas
    games = pd.read_json(archivo_gzip, lines=True)

games
```

```python
items1 = []
# Ruta al archivo comprimido que contiene los datos en formato Gzip
ruta_archivo_comprimido = 'users_items.json.gz'

# Abre el archivo comprimido en modo lectura
with gzip.open(ruta_archivo_comprimido, 'rt', encoding='utf-8') as archivo_gzip:
    # Lee el contenido del archivo Gzip línea por línea
    for line in archivo_gzip:
        # Usa ast.literal_eval para interpretar cada línea como un objeto Python
        item = ast.literal_eval(line)
        items1.append(item)

# Crea un DataFrame de Pandas con los datos
items = pd.DataFrame(items1)
items.head()
```
